#Functions

```python
def image_processing(path, size=(200,200)):
  """
  Función para leer la imagen de una ruta específica y
  redimensionarla en base a un tamaño estándar para
  facilitar su procesamiento vectorial
  Entrada:
  "path" - Ruta del archivo
  "size" - Tamaño estándar planteado (200,200)
  Salida:
  "img" - Imagen procesada
  """
  img = cv2.imread(path, cv2.IMREAD_COLOR_RGB)  #Lectura de la imagen
  img = cv2.resize(img, size) #Escalamiento para menor procesamiento
  return img

def euclidean_distance(img1, img2):
  """
  Función para hallar la distancia euclidea entre
  2 vectores provenientes de las imágenes
  Entrada:
  "img1" - Imagen 1
  "img2" - Imagen 2
  Salida:
  "distance" - Distancia euclidea
  """
  img1 = img1.astype(float).flatten() #Cambiamos a float para evitar desbordamiento
  img2 = img2.astype(float).flatten() #de datos y se aplana en un solo vector
  squared_diffs = [(a - b) ** 2 for a, b in zip(img1, img2)] #Hallamos las diferencias al cuadrado según fórmula
  total = sum(squared_diffs)  #Se realiza la sumatoria de diferencias
  distance = total ** 0.5 #Se halla la raíz de toda la suma
  return distance
```

#Test #1
La idea de esta prueba es tomar 2 imágenes en diferentes estados del actor Ryan Gosling (1,3) y otra de William Shakespare (2)

```python
#Lectura de la ruta de las imágenes desde el Github
imagen1 = '/content/Machine-Learning-Theory/FastWorks/Task1Images/Task1.1.png'
imagen2 = '/content/Machine-Learning-Theory/FastWorks/Task1Images/Task1.2.png'
imagen3 = '/content/Machine-Learning-Theory/FastWorks/Task1Images/Task1.3.png'

#Apertura de las imágenes con CV2 para visualización
img1 = image_processing(imagen1)
img2 = image_processing(imagen2)
img3 = image_processing(imagen3)

#Hallamos distancias euclideas entre cada imagen
d12 = euclidean_distance(img1, img2)
d13 = euclidean_distance(img1, img3)
d23 = euclidean_distance(img2, img3)

#Comparamos distancias entre cada imagen
print(f"Distancia entre imagen 1 e imagen 2: {d12:.2f}")
print(f"Distancia entre imagen 1 e imagen 3: {d13:.2f}")
print(f"Distancia entre imagen 2 e imagen 3: {d23:.2f}")
```

Se puede apreciar bastante similitud entre las imágenes, especialmente la mayor cercanía se da con las fotos del mismo actor. Sin embargo es importante visualizar la cercanía (parecido) entre el actor y William Shakespare, denotado por una pequeña diferencia en el valor de la distancia

```python
plt.figure(figsize=(16,10))
plt.subplot(1,3,1), plt.imshow(img1), plt.title('Imagen 1'), plt.axis('off')
plt.subplot(1,3,2), plt.imshow(img2), plt.title('Imagen 2'), plt.axis('off')
plt.subplot(1,3,3), plt.imshow(img3), plt.title('Imagen 3'), plt.axis('off')
plt.show()
```
