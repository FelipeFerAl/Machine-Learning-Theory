#Libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

#Functions

```python
def image_processing(path, size=(200,200)):
  img = cv2.imread(path, cv2.IMREAD_COLOR_RGB)
  img = cv2.resize(img, size)
  return img

def euclidean_distance(img1, img2):
  # diff = img1 - img2
  # sq_diff = diff**2
  # sum_sq_diff = np.sum(sq_diff)
  # distance = np.sqrt(sum_sq_diff)
  distance = np.linalg.norm(img1 - img2)
  return distance
```

#Main Code

```python
imagen1 = '/content/Tarea1.1.png'
imagen2 = '/content/Tarea1.2.png'
imagen3 = '/content/Tarea1.3.png'

imagen1 = image_processing(imagen1)
imagen2 = image_processing(imagen2)
imagen3 = image_processing(imagen3)

vec1 = imagen1.flatten()
vec2 = imagen2.flatten()
vec3 = imagen3.flatten()

d12 = euclidean_distance(vec1, vec2)
d13 = euclidean_distance(vec1, vec3)

print(f"Distancia entre original e imagen 2: {d12:.2f}")
print(f"Distancia entre original e imagen 3: {d13:.2f}")
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,3,1), plt.imshow(imagen1), plt.title('Original'), plt.axis('off')
plt.subplot(1,3,2), plt.imshow(imagen2), plt.title('Imagen 2'), plt.axis('off')
plt.subplot(1,3,3), plt.imshow(imagen3), plt.title('Imagen 3'), plt.axis('off')
plt.show()
```
